```python
titanic_csv = open('./titanic.csv', 'r')
```

```python
linhas = titanic_csv.readlines()
```

```python
titanic_csv.seek(0)
```

```python
titanic_csv.close()
```

```python
colunas = linhas[0].split(',')
```

```python
colunas
```

```python
titanic_dados = {}
```

```python
for coluna in colunas:
    titanic_dados[coluna] = []
```

```python
dados_como_lista = []

for linha in linhas[1:]:
    dados_como_lista.append(linha.split(','))
```

```python
print(dados_como_lista[0])
```

# dados_como_lista - Tem todos os dados (menos o cabeçalho)
# titanic_dados - Dicionário com o cabeçalho (sem os dados)

```python
print(f'{colunas[3]}: {dados_como_lista[0][3]}')
```

```python
titanic_dados
```

```python
for passageiro in dados_como_lista:
    for k, v in enumerate(passageiro):
        coluna = colunas[k]
        titanic_dados[coluna].append(v)
```

```python
for indice, age in enumerate(titanic_dados['Age']):
    if titanic_dados['Age'][indice] != '':
        titanic_dados['Age'][indice] = float(age)
    else:
        titanic_dados['Age'][indice] = -1
```

```python
idades = []

for idade in titanic_dados['Age']:
    if idade >= 0:
        idades.append(idade)
```

```python
max(titanic_dados['Age'])
```

```python
min(titanic_dados['Age'])
```

```python
total = sum(idades)
media = total / len(idades)
print(media)
```

```python
for indice, age in enumerate(titanic_dados['Age']):
    if titanic_dados['Age'][indice] == -1:
        titanic_dados['Age'][indice] = media
```
