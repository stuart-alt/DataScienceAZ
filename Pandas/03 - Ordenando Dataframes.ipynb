```python
# Ordenar por mais de uma coluna, basta passar uma lista de colunas.
filmes.sort_values(['genre', 'duration']).head(3)
```

```python
# Filtrar por filmes com mais de 180 minutos de duração.
boolean = []

for item in filmes.duration:
    if item >= 180:
        boolean.append(True)
    else:
        boolean.append(False)
```

```python
boolean[:5]   # Imprimndo apenas os 5 primos itens.
```

```python
# Converter a lista em Series
duracao = pd.Series(boolean)
```
